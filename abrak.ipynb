```python
### KEPLER ábra. 2.1
import numpy as np
import math
import matplotlib.pyplot as plt

fi = np.linspace(0,2*np.pi,1000)
a = 14
b = 10
c = math.sqrt(a**2-b**2)

r = []
for i in range (len(fi)):
    r.append(a*b/(math.sqrt(b**2*np.cos(fi[i])**2+a**2*np.sin(fi[i])**2)))

x = r*np.cos(fi)
y = r*np.sin(fi)

tol = 0.026

liney1 = []
liney2 = []


for i in range (1000):
    if abs(x[i] - 12.8) < tol:
        liney1.append(y[i])
    if abs(x[i] - 13.8) < tol:
        liney2.append(y[i])
        

fig = plt.figure()
fig.set_size_inches(15, 10)

plt.plot(x,y, color = 'black', linewidth=3)
plt.scatter(-c,0, color = 'black', s = 100)
plt.arrow(-c, 0, c-12.8, liney1[1], color = 'black',linewidth=3)
plt.arrow(-c, 0, c-12.8, liney1[0], linewidth=3)
plt.arrow(-c, 0, 13.8 + c, liney2[0],linewidth=3)
plt.arrow(-c, 0, 13.8 + c, liney2[3],linewidth=3)
plt.text(-12.8, -0.35, s = '$A_1$', fontsize = 25)
plt.text(5, -0.35, s = '$A_2$', fontsize = 25)
plt.axis('off')

```
